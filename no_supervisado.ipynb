# Aprendizaje no supervisado

Resumen del Aprendizaje No Supervisado

El aprendizaje no supervisado es una rama del aprendizaje automático que se enfoca en descubrir patrones, estructuras y relaciones ocultas en datos sin etiquetas o respuestas predefinidas. A diferencia del aprendizaje supervisado, donde los modelos se entrenan con pares de entrada-salida conocidos, en el aprendizaje no supervisado el modelo debe inferir por sí mismo las estructuras subyacentes en los datos.

Objetivos Principales:

* Descubrir patrones y estructuras ocultas: Identificar agrupaciones naturales, relaciones o estructuras en los datos.
* Reducción de dimensionalidad: Simplificar los datos manteniendo la mayor cantidad posible de información relevante.
* Detección de anomalías: Encontrar datos que no encajan con el patrón general (anomalías o valores atípicos).
* Segmentación: Dividir los datos en grupos significativos para análisis o acciones posteriores.


## Técnicas Comunes del Aprendizaje No Supervisado:

* Clustering (Agrupamiento):

Descripción: Agrupar datos en clusters basados en similitudes o distancias entre ellos.
Algoritmos populares:

K-Means: Divide los datos en k clusters minimizando la variación dentro de cada cluster.
Clustering Jerárquico: Crea una jerarquía de clusters mediante un enfoque aglomerativo (bottom-up) o divisivo (top-down).

* Reducción de Dimensionalidad:

Descripción: Reduce el número de variables (dimensiones) en los datos mientras se conserva la información esencial.
Técnicas populares:
Análisis de Componentes Principales (PCA): Transforma las variables originales en un nuevo conjunto de variables no correlacionadas (componentes principales) ordenadas por la cantidad de varianza explicada.


## Agrupación de Clusters: Cómo Funciona y Técnicas Disponibles

## **Cómo funciona la agrupación de clusters**
La agrupación de clusters es un método de **aprendizaje no supervisado** que agrupa datos no etiquetados en conjuntos o clusters basados en similitudes entre las observaciones. El objetivo principal es que los elementos dentro de un cluster sean más similares entre sí que a los elementos de otros clusters.

### **Pasos básicos del proceso**
1. **Representación de los datos**:
   - Cada observación se representa como un punto en un espacio multidimensional, donde cada dimensión corresponde a una variable del dataset. Si hay más de un par de dimensiones lo común es reducirlas, casi siempre usando PCA


2. **Definición de la similitud**:
   - Se utiliza una métrica de distancia (por ejemplo, distancia Euclidiana, Manhattan o Coseno) para medir qué tan similares son los puntos.



3. **Inicialización de los clusters**:
   - Los algoritmos parten de una hipótesis inicial, como asignar centros de cluster aleatorios.


4. **Asignación de puntos a clusters**:
   - Cada observación se asigna al cluster más cercano según la métrica de similitud definida.


5. **Optimización**:
   - Se ajustan los centroides (o la estructura del cluster) iterativamente para minimizar las distancias internas del cluster. 


6. **Finalización**:
   - El proceso se detiene cuando no hay cambios significativos en los clusters (convergencia) o después de un número fijo de iteraciones.

## Elección del Número de Clusters (\(k\)) en K-Means

La idea es clasificarlos en grupos con caracteristicas comunes (clusters)


### **1. Conocimiento previo del dominio**
   - Si ya conoces las categorías o grupos esperados en los datos, puedes fijar el número de clusters (\(k\)) de antemano.  
   - **Ejemplo**: Si buscas clasificar estudiantes según inteligencias múltiples, puedes fijar \(k = 4\) o \(k = 5\) basándote en categorías previamente definidas.

---

### **2. Propósito del análisis**
   - **Exploratorio**: Para identificar patrones amplios, comienza con \(k = 2\) o \(k = 3\).
   - **Detallado**: Si deseas descubrir diferencias más específicas, elige un valor mayor, como \(k = 10\).

---

### **3. Reglas empíricas comunes**
   - **Raíz cuadrada de \(n\)**:
     \[
     k \approx \sqrt{n/2}
     \]
     - **Ejemplo**: Con 300 observaciones, \(k \approx \sqrt{300/2} \approx 12\).
   - **Fórmula de Sturges** (adaptada para clusters):
     \[
     k \approx 1 + \log_2(n)
     \]
     - **Ejemplo**: Con 300 observaciones, \(k \approx 1 + \log_2(300) \approx 9\).

---

Estas estrategias proporcionan un punto de partida razonable para elegir \(k\) según la naturaleza del problema o el tamaño del dataset.


```python
# Código para crear el dataset y realizar análisis de clustering

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Número de estudiantes por tipología
n_por_tipologia = 100

# Crear listas para cada variable
Puntaje_Logico_Matematico = []
Puntaje_Verbal_Linguistico = []
Puntaje_Visual_Espacial = []
Puntaje_Musical = []
Puntaje_Cinetico_Corporal = []
Puntaje_Interpersonal = []
Puntaje_Intrapersonal = []
Horas_Estudio_Semanal = []
Participacion_En_Clases = []
Preferencia_Grupal = []

# Tipologías
tipologias = ['Matematicos_Razonadores', 'Comunicadores_Sociales', 'Artistas_Creativos', 'Estrategas_Solitarios', 'Generalistas']

# Lista para almacenar las tipologías de cada estudiante
tipologias_list = []

# Generar datos para cada tipología
for tipologia in tipologias:
    for _ in range(n_por_tipologia):
        if tipologia == 'Matematicos_Razonadores':
            # Altas puntuaciones en: Lógico-Matemático (>80), Visual-Espacial (>70)
            # Bajas puntuaciones en: Musical (<50), Intrapersonal (<50)
            Puntaje_Logico_Matematico.append(np.random.uniform(81, 100))
            Puntaje_Visual_Espacial.append(np.random.uniform(71, 100))
            Puntaje_Musical.append(np.random.uniform(10, 49))
            Puntaje_Intrapersonal.append(np.random.uniform(30, 49))

            # Otras puntuaciones aleatorias dentro de sus rangos
            Puntaje_Verbal_Linguistico.append(np.random.uniform(30, 100))
            Puntaje_Cinetico_Corporal.append(np.random.uniform(20, 100))
            Puntaje_Interpersonal.append(np.random.uniform(10, 100))

            # Comportamiento
            Horas_Estudio_Semanal.append(np.random.uniform(20, 40))
            Participacion_En_Clases.append(np.random.uniform(1, 5))
            Preferencia_Grupal.append(0)

        elif tipologia == 'Comunicadores_Sociales':
            # Altas puntuaciones en: Verbal-Lingüístico (>85), Interpersonal (>80)
            # Bajas puntuaciones en: Visual-Espacial (<60), Lógico-Matemático (<50)
            Puntaje_Verbal_Linguistico.append(np.random.uniform(86, 100))
            Puntaje_Interpersonal.append(np.random.uniform(81, 100))
            Puntaje_Visual_Espacial.append(np.random.uniform(20, 59))
            Puntaje_Logico_Matematico.append(np.random.uniform(40, 49))

            # Otras puntuaciones
            Puntaje_Musical.append(np.random.uniform(10, 100))
            Puntaje_Cinetico_Corporal.append(np.random.uniform(20, 100))
            Puntaje_Intrapersonal.append(np.random.uniform(30, 100))

            # Comportamiento
            Horas_Estudio_Semanal.append(np.random.uniform(5, 20))
            Participacion_En_Clases.append(np.random.uniform(6, 10))
            Preferencia_Grupal.append(1)

        elif tipologia == 'Artistas_Creativos':
            # Altas puntuaciones en: Musical (>90), Visual-Espacial (>80), Cinético-Corporal (>75)
            # Baja puntuación en: Lógico-Matemático (<50)
            Puntaje_Musical.append(np.random.uniform(91, 100))
            Puntaje_Visual_Espacial.append(np.random.uniform(81, 100))
            Puntaje_Cinetico_Corporal.append(np.random.uniform(76, 100))
            Puntaje_Logico_Matematico.append(np.random.uniform(40, 49))

            # Otras puntuaciones
            Puntaje_Verbal_Linguistico.append(np.random.uniform(30, 100))
            Puntaje_Interpersonal.append(np.random.uniform(10, 100))
            Puntaje_Intrapersonal.append(np.random.uniform(30, 100))

            # Comportamiento
            Horas_Estudio_Semanal.append(np.random.uniform(5, 25))
            Participacion_En_Clases.append(np.random.uniform(3, 8))
            Preferencia_Grupal.append(np.random.choice([0,1]))

        elif tipologia == 'Estrategas_Solitarios':
            # Altas puntuaciones en: Intrapersonal (>85), Lógico-Matemático (>70)
            # Baja puntuación en: Interpersonal (<50)
            Puntaje_Intrapersonal.append(np.random.uniform(86, 100))
            Puntaje_Logico_Matematico.append(np.random.uniform(71, 100))
            Puntaje_Interpersonal.append(np.random.uniform(10, 49))

            # Otras puntuaciones
            Puntaje_Verbal_Linguistico.append(np.random.uniform(30, 100))
            Puntaje_Visual_Espacial.append(np.random.uniform(20, 100))
            Puntaje_Musical.append(np.random.uniform(10, 100))
            Puntaje_Cinetico_Corporal.append(np.random.uniform(20, 100))

            # Comportamiento
            Horas_Estudio_Semanal.append(np.random.uniform(25, 40))
            Participacion_En_Clases.append(np.random.uniform(1, 5))
            Preferencia_Grupal.append(0)

        elif tipologia == 'Generalistas':
            # Equilibrio en todas las puntuaciones
            Puntaje_Logico_Matematico.append(np.random.uniform(70, 85))
            Puntaje_Verbal_Linguistico.append(np.random.uniform(70, 85))
            Puntaje_Visual_Espacial.append(np.random.uniform(70, 85))
            Puntaje_Musical.append(np.random.uniform(70, 85))
            Puntaje_Cinetico_Corporal.append(np.random.uniform(70, 85))
            Puntaje_Interpersonal.append(np.random.uniform(70, 85))
            Puntaje_Intrapersonal.append(np.random.uniform(70, 85))

            # Comportamiento
            Horas_Estudio_Semanal.append(np.random.uniform(15, 30))
            Participacion_En_Clases.append(np.random.uniform(5, 8))
            Preferencia_Grupal.append(np.random.choice([0,1]))

        # Añadir la tipología a la lista
        tipologias_list.append(tipologia)

# Crear el DataFrame
data = {
    'Puntaje_Lógico_Matemático': Puntaje_Logico_Matematico,
    'Puntaje_Verbal_Lingüístico': Puntaje_Verbal_Linguistico,
    'Puntaje_Visual_Espacial': Puntaje_Visual_Espacial,
    'Puntaje_Musical': Puntaje_Musical,
    'Puntaje_Cinético_Corporal': Puntaje_Cinetico_Corporal,
    'Puntaje_Interpersonal': Puntaje_Interpersonal,
    'Puntaje_Intrapersonal': Puntaje_Intrapersonal,
    'Horas_Estudio_Semanal': Horas_Estudio_Semanal,
    'Participación_En_Clases': Participacion_En_Clases,
    'Preferencia_Grupal': Preferencia_Grupal,
    'Tipología': tipologias_list
}

df = pd.DataFrame(data)

# Mostrar las primeras filas del DataFrame
print("Primeras filas del dataset:")
print(df.head())

# Guardar el DataFrame en un archivo CSV
df.to_csv('dataset_estudiantes.csv', index=False)

# ------------------------------
# Análisis de Clustering
# ------------------------------

# Preparar los datos para el clustering
# Vamos a usar solo las variables numéricas para el clustering
variables_numericas = [
    'Puntaje_Lógico_Matemático',
    'Puntaje_Verbal_Lingüístico',
    'Puntaje_Visual_Espacial',
    'Puntaje_Musical',
    'Puntaje_Cinético_Corporal',
    'Puntaje_Interpersonal',
    'Puntaje_Intrapersonal',
    'Horas_Estudio_Semanal',
    'Participación_En_Clases'
]

X = df[variables_numericas]

# Estandarizar las variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Reducir la dimensionalidad para visualización (opcional)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Determinar el número óptimo de clusters (k) usando el método del codo
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(8, 4))
plt.plot(K, inertia, 'bx-')
plt.xlabel('Número de clusters K')
plt.ylabel('Inercia')
plt.title('El método del codo para encontrar el K óptimo')
plt.show()

# Según el gráfico, elegimos K=5 (ya que tenemos 5 tipologías)
k_optimo = 5

# Aplicar K-Means con K óptimo
kmeans = KMeans(n_clusters=k_optimo, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

# Añadir los clusters al DataFrame
df['Cluster'] = clusters

# Visualizar los clusters en el espacio reducido por PCA
plt.figure(figsize=(8, 6))
for cluster in range(k_optimo):
    plt.scatter(
        X_pca[df['Cluster'] == cluster, 0],
        X_pca[df['Cluster'] == cluster, 1],
        label=f'Cluster {cluster}'
    )
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Clusters de estudiantes (reducción PCA)')
plt.legend()
plt.show()

# Analizar cómo se relacionan los clusters con las tipologías
tabla_contingencia = pd.crosstab(df['Tipología'], df['Cluster'])
print("Tabla de contingencia entre Tipología y Cluster:")
print(tabla_contingencia)

# Visualizar los centroides
centroides = kmeans.cluster_centers_

# Desescalar los centroides
centroides_desescalados = scaler.inverse_transform(centroides)
centroides_df = pd.DataFrame(centroides_desescalados, columns=variables_numericas)

print("\nCentroides de los clusters:")
print(centroides_df)

# Opcional: interpretar los centroides
for i, row in centroides_df.iterrows():
    print(f"\nCaracterísticas del Cluster {i}:")
    for variable in variables_numericas:
        print(f"  {variable}: {row[variable]:.2f}")
```
